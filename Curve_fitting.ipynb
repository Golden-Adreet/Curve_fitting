```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import random
```

```python
def gauss(x, A,B,C,D):
    y = A*x**3 + B*x**2 + C*x + D
    return y

xdata = np.linspace(0,5,100)
ydata = np.array([i**3+random.randrange(-10,10) for i in xdata])
#using Scipy
p,c = curve_fit(gauss,xdata,ydata)
a = p[0]
b = p[1]
c = p[2]
d = p[3]
fiteqnsci = 'Fitted curve\n($'+str(round(a,3))+'x^3 +'+str(round(b,3))+'x^2 +' + str(round(c,3)) +'x + '+ str(round(b,3)) +'$)'
y_fit_scipy = gauss(xdata,a,b,c,d)
#using Numpy
p = np.polyfit(xdata,ydata,3)
fiteqnnp = 'Fitted curve\n($'+str(round(p[0],3))+'x^3 +'+str(round(p[1],3))+'x^2 +' + str(round(p[2],3)) +'x + '+ str(round(p[3],3)) +'$)'
y_fit_numpy = np.polyval(p,xdata)
plt.subplot(2,1,1)
plt.plot(xdata,ydata,'o',xdata,y_fit_scipy)
plt.title('Using Scipy')
plt.legend(['original Data',fiteqnsci],loc='best')
plt.subplot(2,1,2)
plt.plot(xdata,ydata,'o',xdata,y_fit_numpy)
plt.title('Using Numpy')
plt.legend(['original Data',fiteqnnp],loc='best')
plt.show()
```
